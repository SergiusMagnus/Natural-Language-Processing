```python
from transformers import pipeline
from tqdm import tqdm
```

```python
sentence = 'After your workout, remember to focus on maintaining a good water balance.'
```

```python
pipe = pipeline('fill-mask', model='bert-base-uncased')
```

```python
tokens = sentence.split(' ')
```

```python
result = []

for i in range(len(tokens)):
    tokens_copy = tokens.copy()

    tokens_copy[i] = '[MASK]'
    promt = ' '.join(tokens_copy)

    for res in pipe(promt):
        result.append(res['sequence'])

for sentence in tqdm(result.copy()):
    cur_tokens = sentence.split(' ')

    for i in range(1, len(cur_tokens)):
        tokens_copy = cur_tokens.copy()
        tokens_copy.insert(i, '[MASK]')
        promt = ' '.join(tokens_copy)

        for res in pipe(promt):
            result.append(res['sequence'])
```
